                 Mtodo Gauss-Jordan


El mtodo de Gauss-Jordan es una forma ms directa para los sistemas lineales. Ya que con operaciones elementales, se transforma la matriz aumentada del sistema hasta llegar a una forma reducida por filas, lo que permite encontrar la solucin de manera clara y ordenada. En Python, con la ayuda de la librera NumPy se facilita mucho el trabajo, adems de que permite entender bien cada transformacin.
Sistema de ecuaciones que se resuelve:

2x + y  z = 8

3x - y + 2z = -11

2x + y + 2z = 3

El sistema Ax=b se debe resolver transformando la matriz aumentada [Ab] en su forma reducida por filas (RREF) usando el mtodo de Gauss-Jordan, hasta que la matriz de la izquierda se convierta en la matriz identidad I, y la solucin quede en la columna derecha.

PROCEDIMIENTO PARA RESOLVER EN PYTHON
1. Importar la librera NumPy
2. Definir funcin gauss_jordan(A, b):
    a. Convertir A y b a tipo float (para evitar errores con divisiones)
    b. Obtener el nmero de filas (n) de la matriz A
    c. Crear la matriz aumentada: [A | b]
    d. Para cada fila i desde 0 hasta n-1:
        i. Obtener el pivote (elemento A[i][i])
        ii. Si el pivote es 0:
            - Mostrar un error (se necesita hacer reordenamiento o el sistema es singular)
        iii. Dividir toda la fila i por el pivote para hacer que el pivote sea 1
        iv. Para cada otra fila j (j  i):
            - Calcular el factor para anular el elemento en la columna i
            - Restar (factor * fila i) a la fila j, para hacer que ese elemento sea 0
    e. Extraer la solucin del sistema desde la ltima columna de la matriz aumentada
    f. Retornar el vector solucin x
3. En el programa principal:
    a. Definir la matriz A y el vector b
    b. Llamar a la funcin gauss_jordan(A, b)
    c. Imprimir el resultado x

CODIGO PYTHON

```python
import numpy as np

# Funcin para resolver un sistema de ecuaciones usando Gauss-Jordan
def gauss_jordan(A, b):
    A = A.astype(float)  # Asegurar que los valores sean flotantes para evitar problemas con la divisin
    b = b.astype(float)  # Convertir b a tipo float tambin

    n = len(A)  # Nmero de filas (y columnas) de la matriz A, que es una matriz cuadrada
    # Crear la matriz aumentada [A | b], donde b es la columna de resultados
    augmented = np.hstack((A, b.reshape(-1, 1)))  # Usamos np.hstack para agregar b a la matriz A

    # Aplicar eliminacin Gauss-Jordan (transformacin a la forma RREF)
    for i in range(n):
        # Paso 1: Hacer que el pivote (el elemento en la diagonal) sea 1
        pivot = augmented[i][i]  # Tomamos el valor del pivote en la posicin (i, i)
        if pivot == 0:
            # Si el pivote es cero, significa que no se puede hacer la normalizacin. 
            # Esto podra indicar que el sistema es singular, por lo que se lanza un error.
            raise ValueError("Pivote cero encontrado. Reordenamiento necesario.")

        # Dividir toda la fila i por el valor del pivote, para hacer que el pivote sea 1
        augmented[i] = augmented[i] / pivot

        # Paso 2: Hacer que todos los dems elementos en la columna del pivote sean 0
        # Esto garantiza que la matriz se convertir en una forma escalonada reducida por filas (RREF)
        for j in range(n):
            if i != j:  # No necesitamos cambiar el valor de la fila i (porque ya est normalizada)
                factor = augmented[j][i]  # Tomamos el valor en la fila j, columna i (lo que queremos hacer 0)
                # Restamos el factor multiplicado por la fila i a la fila j
                augmented[j] = augmented[j] - factor * augmented[i]

    # Una vez que hemos terminado, la matriz aumentada tiene la forma [I | x]
    # La solucin est ahora en la ltima columna (al lado derecho de la matriz aumentada)
    x = augmented[:, -1]  # La ltima columna de la matriz es la solucin del sistema

    return x

# Ejemplo

# Matriz de coeficientes A (3x3)
A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]], dtype=float)

# Vector de trminos independientes b (3x1)
b = np.array([8, -11, -3], dtype=float)

# Resolver el sistema usando Gauss-Jordan
x = gauss_jordan(A, b)

# Mostrar la solucin
print("Solucin del sistema (Gauss-Jordan):\n", x)
```

COMENTARIO FINAL

El enfoque de Gauss-Jordan es un procedimiento para solucionar sistemas de ecuaciones lineales a travs de operaciones elementales en la matriz incrementada.  En contraposicin al mtodo de Gauss (eliminacin sencilla), Gauss-Jordan prosigue con el procedimiento hasta alcanzar la forma escalonada reducida por filas (RREF), lo que facilita la localizacin de la solucin directamente sin la necesidad de reemplazos subsiguientes.  A pesar de que no siempre es el ms eficaz en computacin para sistemas de gran envergadura.
