                 Métodos de Sistemas Lineales


INTRODUCCIÓN

Cuando se requiere resolver sistemas de ecuaciones lineales, es importante conocer las diversas formas y procedimientos que existen para resolverlos con mayor eficacia dependiendo de sus características, estos son indispensables en el álgebra lineal además de tener usos directos en varios campos de la ciencia, la ingeniería y la economía. Estos procedimientos no solo facilitan la resolución rápida de sistemas, además que se abordan diversas visiones acerca de la estructura de las matrices y el funcionamiento de los sistemas lineales. Cada método posee sus pros, contras y situaciones ideales de aplicación.  En esta ocasión se trabajará con cuatro de los métodos más destacados: la matriz inversa, la descomposición LU, el método de Gauss-Jordan y el método del máximo elemento pivote.
